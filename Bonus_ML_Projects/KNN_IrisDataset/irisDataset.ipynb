```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn import preprocessing
from sklearn.neighbors import KNeighborsClassifier
from sklearn import metrics
```

```python
irisDataset=pd.read_csv('iris.csv')
```

dataset count

```python
irisDataset['variety'].value_counts()
```

encode verity string for traing

```python
y=irisDataset[['variety']].values.ravel()
label_encoder=LabelEncoder()
#covert string to number
Y=label_encoder.fit_transform(y)
```

```python
X=irisDataset[['sepal.length', 'sepal.width', 'petal.length', 'petal.width']].values
```

```python
# create test and train data
x_train,x_test,y_train,y_test=train_test_split(X,Y,test_size=0.2,random_state=4)
```

normalize the values and spread from large number devation to mean number devation 
example if number range from 0 to 100 then normalization will find the mean and then normalize whole data between smaller range like from 0 to 5

```python
X_train_norm = preprocessing.StandardScaler().fit(x_train).transform(x_train.astype(float))
X_test_norm = preprocessing.StandardScaler().fit(x_test).transform(x_test.astype(float))
```

running KNN

```python
# intitalizing k as 4
k=4
```

```python
#Train Model and Predict
neigh = KNeighborsClassifier(n_neighbors = k).fit(X_train_norm,y_train)
```

```python
yhat = neigh.predict(X_test_norm)
```

```python
print("Train set Accuracy: ", metrics.accuracy_score(y_train, neigh.predict(X_train_norm)))
print("Test set Accuracy: ", metrics.accuracy_score(y_test, yhat))
```

# check variable k value and see whiich gives higest score

```python
best_accuracy=0
best_k=0
for k in range(1,120):
    neigh = KNeighborsClassifier(n_neighbors = k).fit(X_train_norm,y_train)
    yhat = neigh.predict(X_test_norm)
    train_accuracy=metrics.accuracy_score(y_train, neigh.predict(X_train_norm))
    test_accuracy=metrics.accuracy_score(y_test, yhat)
    print('For k=',k,'accurachy is',train_accuracy,test_accuracy)
    if(test_accuracy>best_accuracy):
        best_accuracy=test_accuracy
        best_k=k

print('print best k with accuracy is',best_k,best_accuracy)
    
```
