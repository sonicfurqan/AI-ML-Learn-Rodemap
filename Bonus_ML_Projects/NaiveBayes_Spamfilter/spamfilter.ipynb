#  using sklearn to predecti spam

```python
def process(text):
    # lowercase it
    text = text.lower()
    
    # return token list
    return text
```

```python
spamDataset=pd.read_csv('spam.csv',encoding='latin-1')[['v1', 'v2']]
spamDataset.columns = ['label', 'message']
```

```python

spam_filter = Pipeline([
    ('vectorizer', TfidfVectorizer(analyzer=process)), # messages to weighted TFIDF score
    ('classifier', MultinomialNB())                    # train on TFIDF vectors with Naive Bayes
])
```

```python
x_train, x_test, y_train, y_test = train_test_split(spamDataset['message'], spamDataset['label'], test_size=0.20, random_state = 21)
```

```python
# train dataset
spam_filter.fit(x_train, y_train)
```

```python
predictions = spam_filter.predict(x_test)
```

```python
count = 0
for i in range(len(y_test)):
    if y_test.iloc[i] != predictions[i]:
        count += 1
print('Total number of test cases', len(y_test))
print('Number of wrong of predictions', count)
```

```python
from sklearn.metrics import confusion_matrix
```

```python
# Generate the matrix
# y_test = actual values, y_pred = predicted values from RFE
cm = confusion_matrix(y_test, predictions)

# Extract TP, TN, FP, FN
# .ravel() flattens the 2x2 matrix into a single line
tn, fp, fn, tp = cm.ravel()

print(f"True Negatives: {tn}")
print(f"False Positives: {fp}")
print(f"False Negatives: {fn}")
print(f"True Positives: {tp}")
```
